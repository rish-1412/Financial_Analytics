```python
# Sort the dataset by Market_Capitalization in descending order
data_sorted_mar_cap = data.sort_values(by='Mar Cap - Crore', ascending=False)

# Select the top 10 companies
top_10_companies = data_sorted_mar_cap.head(10)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.barh(top_10_companies['Company Name'], top_10_companies['Mar Cap - Crore'], color='skyblue')
plt.xlabel('Market Capitalization (in crores)')
plt.ylabel('Company Name')
plt.title('Top 10 Companies by Market Capitalization')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value on top
plt.show()
```

```python
# Sort the dataset by Quarterly sales revenue in descending order
data_sorted_sales = data.sort_values(by='Sales Qtr - Crore', ascending=False)

# Select the top 10 companies
top_10_company_sales = data_sorted_sales.head(10)
```

```python
# Plotting the line chart
plt.figure(figsize=(20, 8))
plt.plot(top_10_company_sales['Company Name'], top_10_company_sales['Sales Qtr - Crore'], marker='o', linestyle='-', color='green', label= top_10_company_sales['Sales Qtr - Crore'])

# Adding titles and labels
plt.title('Top 10 Companies By Sales Revenue')
plt.xlabel('Company Name')
plt.ylabel('Quarterly Sales Revenue (in crores)')
plt.grid(True)  # Add a grid for better readability

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Dropping rows with value as 0
data = data[(data != 0).all(axis=1)]

# Sort the dataset by 'Market_Capitalization' in ascending order
data_asc_sorted = data.sort_values(by='Mar Cap - Crore', ascending=True)

# Select the bottom 10 companies
bottom_10_companies = data_asc_sorted.head(15)
```

```python
# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(bottom_10_companies['Mar Cap - Crore'], 
        labels=bottom_10_companies['Company Name'], 
        autopct=lambda p: '{:.0f}'.format(p * sum(bottom_10_companies['Mar Cap - Crore']) / 100), 
        startangle=140)

# Adding title
plt.title('Bottom 15 Companies by Market Capitalization')

# Display the pie chart
plt.show()
```

```python
# Sort the dataset by 'Quarterly Sales Revenue' in ascending order
data_asc_sorted_sales = data.sort_values(by='Sales Qtr - Crore', ascending=True)

# Select the bottom 10 companies
bottom_10_company_sales = data_asc_sorted.head(15)
```

```python
# Plotting the chart
plt.figure(figsize=(10, 8))
sns.barplot(x=bottom_10_company_sales["Sales Qtr - Crore"], y=bottom_10_company_sales["Company Name"], palette= "pastel")
plt.xlabel("Quarterly Sales Revenue")
plt.ylabel("Company Name")
plt.title("Bottom 15 Companies by Quarterly Sales Revenue")
plt.show()
```

```python
# Sort by Market Capitalization and select top 3 companies
df_sorted = data.sort_values(by='Mar Cap - Crore', ascending=False).head(3)

# Plot Bar Chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.2

# Positions for the bars
bar_positions1 = range(len(df_sorted['Company Name']))
bar_positions2 = [p + bar_width for p in bar_positions1]

# Plotting Market Capitalization
ax.bar(bar_positions1, df_sorted['Mar Cap - Crore'], bar_width, label='Mar Cap - Crore', color='orange')

# Plotting Quarterly Sales Revenue
ax.bar(bar_positions2, df_sorted['Sales Qtr - Crore'], bar_width, label='Sales Qtr - Crore', color='green')

# Plotting the chart
ax.set_xlabel('Company Name')
ax.set_ylabel('Value (in Crores)')
ax.set_title('Top 3 Companies by Market Capitalization vs Quarterly Sales Revenue')
ax.set_xticks([p + bar_width / 2 for p in bar_positions1])
ax.set_xticklabels(df_sorted['Company Name'])
ax.legend()

plt.show()
```

```python
# Sort by Market Capitalization and select bottom 3 companies
df_sorted_1 = data.sort_values(by='Mar Cap - Crore', ascending=True).head(3)

# Plot Bar Chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width_ = 0.2

# Positions for the bars
bar_positions_1 = range(len(df_sorted_1['Company Name']))
bar_positions_2 = [p + bar_width_ for p in bar_positions_1]

# Plotting Market Capitalization
ax.bar(bar_positions_1, df_sorted_1['Mar Cap - Crore'], bar_width_, label='Mar Cap - Crore', color='y')

# Plotting Quarterly Sales Revenue
ax.bar(bar_positions_2, df_sorted_1['Sales Qtr - Crore'], bar_width_, label='Sales Qtr - Crore', color='b')

# Plotting the chart
ax.set_xlabel('Company Name')
ax.set_ylabel('Value (in Crores)')
ax.set_title('Bottom 3 Companies by Market Capitalization vs Quarterly Sales Revenue')
ax.set_xticks([p + bar_width_ / 2 for p in bar_positions_1])
ax.set_xticklabels(df_sorted_1['Company Name'])
ax.legend()

plt.show()
```
